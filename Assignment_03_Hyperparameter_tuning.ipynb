##  Feature Scaling of Numerical Attributes

```python
X = df.drop(['DEFAULT_PAYMENT'],axis=1)   # features
y = df['DEFAULT_PAYMENT']               # target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()# scaling
X = scaler.fit_transform(X)
```

##  Spiliting Dataset into training(70%) and test set(30%)

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Modelling:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer
```

```python
# Initialize and train different models
models = {
    "Logistic Regression" : LogisticRegression(),
    'kNN': KNeighborsClassifier(),
    'SVM': SVC(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'XGBoost': XGBClassifier()
}

results = []

for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    
    results.append([model_name, accuracy, f1, precision, recall, roc_auc])

# Print the results
columns = ['Model', 'Accuracy', 'F1 Score', 'Precision', 'Recall', 'ROC AUC']
result_df = pd.DataFrame(results, columns=columns)
print(result_df)
```

##  Model Optimization using K-Fold Cross validation , Grid Search and Hyperopt 

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
model_mean_accuracy=[]
model_std=[]

# Initialize models
lg_model = LogisticRegression()
knn_model = KNeighborsClassifier()
svm_model = SVC()
gb_model = GradientBoostingClassifier()
ada_model = AdaBoostClassifier()
xg_model = XGBClassifier()
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

# Assuming you have defined lg_model, knn_model, svm_model, gb_model, ada_model, xg_model
classifiers = ['Logistic Regression', 'KNN', 'SVM', 'Gradient Boosting', 'AdaBoost', 'XGBoost']
models = [lg_model, knn_model, svm_model, gb_model, ada_model, xg_model]

model_mean_accuracy = []
model_std = []
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for classifier, model in zip(classifiers, models):
    cv_result = cross_val_score(model, X, y, cv=kfold, scoring="accuracy",n_jobs=-1)
    model_mean_accuracy.append(np.mean(cv_result))
    model_std.append(np.std(cv_result))
```

```python
stratified_k_fold_performance = pd.DataFrame({'Mean Accuracy': model_mean_accuracy, 'Std': model_std}, index=classifiers)
print(stratified_k_fold_performance)
```

## Hyperopt

```python
from hyperopt import fmin, tpe, hp
# Hyperopt Logistic Regression
# Define the objective function
def objective(params):
    model = LogisticRegression(**params)
    score = cross_val_score(model, X, y, cv=5, scoring='accuracy').mean()
    return -score  # Hyperopt minimizes, so we use negative accuracy
```

```python
# Define the search space
space = {
    'C': hp.loguniform('C', -3, 3),  # Search for C values in a log scale
    'penalty': hp.choice('penalty', ['l2', 'none']),  # Use 'l2' or 'none' penalty
}
```

```python
# Run Hyperopt
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=10,  # Adjust the number of evaluations
            verbose=1)

print("Best hyperparameters:", best)
```
